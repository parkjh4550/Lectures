```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, random_state = 100)
```

```python
# 어떤 클러스터에 속하는 데이터인지 모른다고 가정
sns.scatterplot(x=data[0][:,0], y=data[0][:,1])
# sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=data[1], palette='Set2')  # 이상적인 클러스터링 상태
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(data[0])
```

```python
model.labels_
```

```python
data[1]
```

```python
sns.scatterplot(x=data[0][:,0], y=data[0][:,1], hue=model.labels_, palette='Set2')
```

```python
distance = []
for i in range(2, 11):
  model = KMeans(n_clusters=i)
  model.fit(data[0])
  distance.append(model.inertia_)
```

```python
distance
```

```python
sns.lineplot(x=list(range(2,11)), y=distance)
```

```python

distance = []
for i in range(2, 11):
  model = KMeans(n_clusters=i)
  model.fit(data[0])
  distance.append(silhouette_score(data[0], model.labels_))
```

```python
sns.lineplot(x=list(range(2,11)), y=distance) # 단순 inertia를 이용할때보다 더 정확한 결과를 얻을 수 있다.
```
