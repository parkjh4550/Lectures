```python
# Standard Scaler, Robust Scaler, MinMax Scaler를 연습
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data load
pd.set_option('display.max_columns', 30)
data_path = '/content/drive/MyDrive/개인 연구/강의 수강/(패스트 캠퍼스) 데이터 분석/Ch4 KNN/churn.csv'
data = pd.read_csv(data_path)
```

```python
# Categorical를 Numeric 변수로 변환
col_list = []
for col in data.columns:
  if data[col].dtype == 'O':
    col_list.append(col)

col_list = col_list[1:] # ID의 경우는 의미가 없으므로 제거
data = pd.get_dummies(data, columns=col_list, drop_first=True)
data.drop('customerID', axis=1, inplace=True)    # customerID column을 제거
data.head(3)
```

```python
# 1. Standard Scaler
# mean 값을 이용하기에 outlier의 영향을 많이 받음
standard = StandardScaler()
standard.fit(data.drop('Churn_Yes', axis=1))    # 종속변수인 column은 제외하고 scaling
```

```python
scaled_st = standard.transform(data.drop('Churn_Yes', axis=1))
```

```python
pd.DataFrame(scaled_st, columns=data.drop('Churn_Yes', axis=1).columns)
```

```python
# 2. Robust Scaler
# Standard Scaler에 비해 outlier의 영향을 덜 받음
```

```python
rob = RobustScaler()
rob.fit(data.drop('Churn_Yes', axis=1))
```

```python
scaled_rob = rob.transform(data.drop('Churn_Yes', axis=1))
```

```python
pd.DataFrame(scaled_rob, columns=data.drop('Churn_Yes', axis=1).columns)
```

```python
# 3. Min-Max Scaler
# Standard, Robust Scaler에 비해 원본 데이터의 분포를 유지하면서 Scaling 가능 -> 가장 무난 사용 가능
# 좀 더 정규화를 필요하면 Standard, Robust Scaler를 사용
minmax = MinMaxScaler()
minmax.fit(data)
print(minmax.n_samples_seen_,  minmax.feature_range)  # 샘플 개수와 scale 값 범위
print(minmax.data_min_, minmax.data_max_) # 각 column별 데이터의 최소, 최대값
```

```python
scaled_data = minmax.transform(data)    # Scaling 결과 : numpy 값으로 결과가 나온다. 
```

```python
pd.DataFrame(scaled_data, columns=data.columns)   # numpy를 dataFrame에 다시 넣어주기
```
