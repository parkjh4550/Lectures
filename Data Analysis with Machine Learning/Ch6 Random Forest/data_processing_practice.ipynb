DataFrame merge

```python
df1 = pd.DataFrame({'id':['X01', 'X02', 'X03', 'X04', 'X05'],
                    'A':[1,3,5,7,9],
                    'B':[11,12,13,14,15],
                    'C':[0,1,2,3,4],
                    'D':[101,102,103,104,105]})
df2 = pd.DataFrame({'id':['X01', 'X02', 'X03', 'X04', 'X07'],
                    'E':[2,4,6,8,10],
                   'F':[1,2,3,4,5],
                   'G':[30,31,32,33,34],
                   'H':[1011,1012,1013,1014,1015]})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='id')   # df1, df2의 교집합 되는 데이터만 merge
# pd.merge(df1, df2, on='id', how='inner')    # 위의 식과 동치
```

```python
pd.merge(df1, df2, on='id', how='outer')  # df1, df2의 합집합 시 얻어지는 데이터들을 모두 포함하여 merge
```

```python
pd.merge(df1, df2, on='id', how='left') # df1에 해당하는 데이터만 merge   # X05가 포함됨
```

```python
pd.merge(df1, df2, on='id', how='right') # df2에 해당하는 데이터만 merge  # X07가 포함됨
```

```python
# ID 이름이 다르게 지정된 경우
df1 = pd.DataFrame({'id':['X01', 'X02', 'X03', 'X04', 'X05'],
                    'A':[1,3,5,7,9],
                    'B':[11,12,13,14,15],
                    'C':[0,1,2,3,4],
                    'D':[101,102,103,104,105]})
df2 = pd.DataFrame({'user_id':['X01', 'X02', 'X03', 'X04', 'X07'],
                    'E':[2,4,6,8,10],
                   'F':[1,2,3,4,5],
                   'G':[30,31,32,33,34],
                   'H':[1011,1012,1013,1014,1015]})
```

```python
# 각각의 id이름을 명시하여 merge
pd.merge(df1, df2, left_on='id', right_on='user_id')
```

DataFrame Join

```python
df1 = pd.DataFrame({'A':[1,3,5,7,9],
                    'B':[11,12,13,14,15],
                    'C':[0,1,2,3,4],
                    'D':[101,102,103,104,105]}, index=[1,2,3,4,5])
df2 = pd.DataFrame({'E':[2,4,6,8,10],
                   'F':[1,2,3,4,5],
                   'G':[30,31,32,33,34],
                   'H':[1011,1012,1013,1014,1015]}, index=[6,7,8,9,10])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2)
```

```python
df1.reset_index(drop=True, inplace=True)
df2.reset_index(drop=True, inplace=True)
```

```python
df1.join(df2)
```

```python
df1.join(df2, how='right')
```
